```python
import matplotlib.pyplot as plt
from scipy.stats import linregress

# =================================================================
# PART 1: Data Conversion and Calculation (mg/L to meq/L)
# =================================================================

# Equivalent Weights (E.W. = Atomic Weight / Valence)
EW = {
    'ca': 20.04,
    'mg': 12.16,
    'k': 39.10,
    'cl': 35.45,
    'hco3': 61.02,
    'so4': 48.03
}

# Convert major ions to meq/L using the standardized, lowercase names
df['Ca_meq'] = df['ca'] / EW['ca']
df['Mg_meq'] = df['mg'] / EW['mg']
df['K_meq'] = df['k'] / EW['k']
df['Cl_meq'] = df['cl'] / EW['cl']
df['HCO3_meq'] = df['hco3'] / EW['hco3']
df['SO4_meq'] = df['so4'] / EW['so4']

# Calculate the specific sums requested for Plot 1
df['Cations_Sum'] = df['Ca_meq'] + df['Mg_meq']
df['Anions_Sum'] = df['HCO3_meq'] + df['SO4_meq']


# =================================================================
# PART 2: Plot Generation
# =================================================================

# Define a function to apply consistent styling (bold black frame, etc.)
def set_plot_style(ax, title, xlabel, ylabel, limit=None):
    ax.set_title(title, fontsize=14)
    ax.set_xlabel(xlabel, fontsize=12)
    ax.set_ylabel(ylabel, fontsize=12)
    ax.grid(True, linestyle=':', alpha=0.6)

    for spine in ['top', 'bottom', 'left', 'right']:
        ax.spines[spine].set_color('black')
        ax.spines[spine].set_linewidth(1.5)

    if limit is not None:
        ax.set_xlim(0, limit)
        ax.set_ylim(0, limit)
        ax.set_aspect('equal', adjustable='box')


# --- Plot 1: (Ca+Mg) vs (HCO₃ + SO₄) in meq/L (1:1 Line) ---
fig, ax1 = plt.subplots(figsize=(7, 7))

# Calculate max limit for the 1:1 line
max_val_1 = max(df['Cations_Sum'].max(), df['Anions_Sum'].max()) * 1.1
ratio_coords_1 = [0, max_val_1]

ax1.scatter(df['Cations_Sum'], df['Anions_Sum'], label='Groundwater Samples', color='green', s=50)
ax1.plot(ratio_coords_1, ratio_coords_1, linestyle='--', color='black', linewidth=1.5, label='1:1 Ratio Line')

set_plot_style(ax1,
               'Cations vs Anions Partial Balance',
               r'$\mathrm{Ca^{2+}} + \mathrm{Mg^{2+}} \ (\mathrm{meq/L})$',
               r'$\mathrm{HCO_3^-} + \mathrm{SO_4^{2-}} \ (\mathrm{meq/L})$',
               limit=max_val_1)
ax1.legend(loc='upper left')
plt.tight_layout()
plt.show()


# --- Plot 2: K⁺ vs Cl⁻ in meq/L (1:1 Line) ---
fig, ax2 = plt.subplots(figsize=(7, 7))

# Calculate max limit
max_val_2 = max(df['K_meq'].max(), df['Cl_meq'].max()) * 1.1
ratio_coords_2 = [0, max_val_2]

ax2.scatter(df['K_meq'], df['Cl_meq'], label='Groundwater Samples', color='red', s=50)
ax2.plot(ratio_coords_2, ratio_coords_2, linestyle='--', color='gray', linewidth=1.5, label='1:1 Ratio Line')

set_plot_style(ax2,
               'K⁺ vs Cl⁻ Concentration (meq/L)',
               r'$\mathrm{K^{+}} \ (\mathrm{meq/L})$',
               r'$\mathrm{Cl^{-}} \ (\mathrm{meq/L})$',
               limit=max_val_2)
ax2.legend(loc='upper left')
plt.tight_layout()
plt.show()


# --- Plot 3: EC vs TDS (Linear Trend Line) ---
fig, ax3 = plt.subplots(figsize=(7, 7))

# Note: Using the standardized 'ec' and 'tds' columns
if 'ec' in df.columns and 'tds' in df.columns:
    # Calculate linear regression (TDS = k * EC + b)
    slope, intercept, r_value, p_value, std_err = linregress(df['ec'], df['tds'])
    r_squared = r_value**2

    # Generate the best-fit line data points
    p = np.poly1d([slope, intercept])
    x_fit = np.linspace(df['ec'].min(), df['ec'].max(), 100)

    # Scatter plot
    ax3.scatter(df['ec'], df['tds'], label='Groundwater Samples', color='purple', s=50)

    # Regression line
    line_label = f'$TDS = {slope:.2f} \cdot EC + {intercept:.2f} \quad (R^2 = {r_squared:.3f})$'
    ax3.plot(x_fit, p(x_fit), color='black', linestyle='-', label=line_label)

    set_plot_style(ax3,
                   'EC vs TDS Correlation',
                   r'Electrical Conductivity ($\mathrm{\mu S/cm}$ or $\mathrm{mS/cm}$)',
                   r'Total Dissolved Solids ($\mathrm{mg/L}$)')
    ax3.legend(loc='lower right')
else:
    ax3.text(0.5, 0.5, 'EC or TDS data columns not found.', ha='center', va='center', color='red')
    set_plot_style(ax3, 'EC vs TDS Correlation', 'EC', 'TDS')

plt.tight_layout()
plt.show()

print("All three plots have been successfully generated.")
```

```python
import pandas as pd

# ------------------------------------------
# 1. Create DataFrame from pasted values
# ------------------------------------------
data = {
    "T.C": [8,9,10,5,9,11,5,10,9,11,10,6,7,9,5,9,6,8,8,3,10,7,9,5,7],
    "T.M.c": [200,190,200,190,170,250,300,270,240,300,180,200,220,270,300,200,260,190,300,290,230,190,169,188,288]
}

df = pd.DataFrame(data)

# ------------------------------------------
# 2. TC classification
# ------------------------------------------
def classify_tc(tc):
    if tc <= 20:
        return "Very Low"
    elif tc <= 100:
        return "Low"
    elif tc <= 1000:
        return "Moderate"
    elif tc <= 10000:
        return "High"
    else:
        return "Very High"

df["TC_Class"] = df["T.C"].apply(classify_tc)

# ------------------------------------------
# 3. TMC classification
# ------------------------------------------
def classify_tmc(tmc):
    if tmc < 100:
        return "Acceptable"
    elif tmc <= 500:
        return "Moderate"
    elif tmc <= 5000:
        return "High"
    else:
        return "Very High"

df["TMC_Class"] = df["T.M.c"].apply(classify_tmc)

# ------------------------------------------
# 4. MPI Index
# ------------------------------------------
df["MPI"] = (df["T.C"] / 100) + (df["T.M.c"] / 500)

def classify_mpi(mpi):
    if mpi <= 1:
        return "Class I – Very Low Pollution"
    elif mpi <= 2:
        return "Class II – Low Pollution"
    elif mpi <= 4:
        return "Class III – Moderate Pollution"
    else:
        return "Class IV – High Pollution"

df["MPI_Class"] = df["MPI"].apply(classify_mpi)

# ------------------------------------------
# 5. EHRC
# ------------------------------------------
def classify_ehrc(row):
    tc = row["T.C"]
    tmc = row["T.M.c"]

    if tc < 20 and tmc < 300:
        return "Very Low Risk"
    elif tc < 100 and tmc < 500:
        return "Low Risk"
    elif (100 <= tc <= 1000) or (500 <= tmc <= 5000):
        return "Moderate Risk"
    else:
        return "High Risk"

df["EHRC"] = df.apply(classify_ehrc, axis=1)

# ------------------------------------------
# 6. Show final classified table
# ------------------------------------------
print(df)
```

```python
import pandas as pd
import numpy as np

# 1. Input Data
# T.C is assumed to be Microbe 1 Concentration (C1)
# T.M.C is assumed to be Microbe 2 Concentration (C2)
data = {
    'T.C': [8, 9, 10, 5, 9, 11, 5, 10, 9, 11, 10, 6, 7, 9, 5, 9, 6, 8, 8, 3, 10, 7, 9, 5, 7],
    'T.M.C': [200, 190, 200, 190, 170, 250, 300, 270, 240, 300, 180, 200, 220, 270, 300, 200, 260, 190, 300, 290, 230, 190, 169, 188, 288]
}
df = pd.DataFrame(data)

# -----------------------------------------------------------
# 2. CODE FOR MPI (Microbiological Pollution Index)
# MPI is calculated as the Geometric Mean: MPI = sqrt(T.C * T.M.C)
# -----------------------------------------------------------
df['MPI_Index'] = np.sqrt(df['T.C'] * df['T.M.C'])

print("--- MPI Calculation Result (First 5 Rows) ---")
print(df[['T.C', 'T.M.C', 'MPI_Index']].head())
print("-" * 40)


# -----------------------------------------------------------
# 3. CODE FOR RISK CLASSIFICATION
# Classification is based on concentration ranges.
# **NOTE**: Medium and High thresholds are HYPOTHETICAL and must be verified
# -----------------------------------------------------------

# Define the risk thresholds
LOW_TC_MIN, LOW_TC_MAX = 5, 11
LOW_TMC_MIN, LOW_TMC_MAX = 150, 288
HIGH_TC_MIN = 20    # Hypothetical
HIGH_TMC_MIN = 400  # Hypothetical

def classify_microbial_risk(tc_conc, tmc_conc):
    """Classifies risk based on concentration ranges."""

    # 1. High Risk Check (If either concentration exceeds severe limit)
    if tc_conc > HIGH_TC_MIN or tmc_conc > HIGH_TMC_MIN:
        return 'High'

    # 2. Low Risk Check (If both concentrations are within the specified safe range)
    is_low_tc = (tc_conc >= LOW_TC_MIN) and (tc_conc <= LOW_TC_MAX)
    is_low_tmc = (tmc_conc >= LOW_TMC_MIN) and (tmc_conc <= LOW_TMC_MAX)

    if is_low_tc and is_low_tmc:
        return 'Low'

    # 3. Medium Risk (Any case that is NOT Low and NOT High)
    return 'Medium'

# Apply the classification function to the DataFrame
df['Risk_Class'] = df.apply(lambda row: classify_microbial_risk(row['T.C'], row['T.M.C']), axis=1)

print("\n--- Risk Classification Result (First 5 Rows) ---")
print(df[['T.C', 'T.M.C', 'Risk_Class']].head())
print("-" * 40)
print(f"Risk Distribution:\n{df['Risk_Class'].value_counts()}")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# T.C and T.M.C values (CFU/mL)
Tc = [8,9,10,5,9,11,5,10,9,11,10,6,7,9,5,9,6,8,8,3,10,7,9,5,7]
Tmc = [200,190,200,190,170,250,300,270,240,300,180,200,220,270,300,200,260,190,300,290,230,190,169,188,288]

# Create dataframe
df = pd.DataFrame({
    'T.C': Tc,
    'T.M.C': Tmc
})

# Calculate MPI
df['MPI'] = np.sqrt((df['T.C']**2 + df['T.M.C']**2)/2)

# Risk classification
def classify_risk(mpi):
    if mpi <= 500:
        return 'Low'
    elif mpi <= 1500:
        return 'Medium'
    else:
        return 'High'

df['Risk'] = df['MPI'].apply(classify_risk)

# Show dataframe
print(df)

# Boxplot of MPI by Risk
plt.figure(figsize=(8,5))
sns.boxplot(x='Risk', y='MPI', data=df, palette='Set1', order=['Low','Medium','High'])
plt.title("Microbial Pollution Index (MPI) by Risk Level")
plt.ylabel("MPI (CFU/mL)")
plt.show()
```
