```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set(style="whitegrid", palette="pastel")
```

```python
np.random.seed(42)  # لضمان تكرار نفس القيم

n_samples = 50

data = {
    'Age_Group': ['Children']*n_samples + ['Adults']*n_samples,
    'Oral_Exposure_mg_per_day': np.concatenate([
        np.random.normal(loc=0.5, scale=0.1, size=n_samples),
        np.random.normal(loc=1.2, scale=0.2, size=n_samples)
    ]),
    'Dermal_Exposure_mg_per_day': np.concatenate([
        np.random.normal(loc=0.2, scale=0.05, size=n_samples),
        np.random.normal(loc=0.5, scale=0.1, size=n_samples)
    ])
}

df = pd.DataFrame(data)
df.head(10)
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Age_Group', y='Dermal_Exposure_mg_per_day', data=df, palette='Set3')
plt.title('Daily Dermal Exposure to Water (mg/day)')
plt.ylabel('Dermal Exposure (mg/day)')
plt.xlabel('Age Group')
plt.show()
```

```python
import pandas as pd
from google.colab import files
import numpy as np

# 1. Upload the XLSX file (a file picker will appear)
print("Please upload your Excel file (.xlsx) now.")
uploaded = files.upload()
file_name = next(iter(uploaded))

# 2. Read the uploaded Excel file into a Pandas DataFrame
try:
    df = pd.read_excel(uploaded[file_name], engine='openpyxl')
except Exception as e:
    print(f"❌ Error loading Excel file: {e}. Please ensure it is a valid .xlsx file.")

# 3. Standardize and Clean Column Names
# This fixes casing issues (like 'Ec' vs 'ec') and removes hidden spaces.
df.columns = df.columns.str.strip()

# Rename columns to standardized lowercase names used in the plotting code:
df = df.rename(columns={
    'Ec': 'ec',
    'TDS': 'tds',
    'Ca': 'ca',
    'Mg': 'mg',
    'K': 'k',
    'CL': 'cl', # Ensures 'CL' is converted to 'cl'
    'HCO3': 'hco3',
    'SO4': 'so4'
}, errors='ignore')

# Display the first few rows to verify the successful load
print("\n✅ Data loaded and columns standardized.")
print("--- Standardized Column Names ---")
print(df.columns.tolist())
print("\nFirst 5 rows of data:")
print(df.head())
```

```python
import matplotlib.pyplot as plt
from scipy.stats import linregress

# =================================================================
# PART 1: Data Conversion and Calculation (mg/L to meq/L)
# =================================================================

# Equivalent Weights (E.W. = Atomic Weight / Valence)
EW = {
    'ca': 20.04,
    'mg': 12.16,
    'k': 39.10,
    'cl': 35.45,
    'hco3': 61.02,
    'so4': 48.03
}

# Convert major ions to meq/L using the standardized, lowercase names
df['Ca_meq'] = df['ca'] / EW['ca']
df['Mg_meq'] = df['mg'] / EW['mg']
df['K_meq'] = df['k'] / EW['k']
df['Cl_meq'] = df['cl'] / EW['cl']
df['HCO3_meq'] = df['hco3'] / EW['hco3']
df['SO4_meq'] = df['so4'] / EW['so4']

# Calculate the specific sums requested for Plot 1
df['Cations_Sum'] = df['Ca_meq'] + df['Mg_meq']
df['Anions_Sum'] = df['HCO3_meq'] + df['SO4_meq']


# =================================================================
# PART 2: Plot Generation
# =================================================================

# Define a function to apply consistent styling (bold black frame, etc.)
def set_plot_style(ax, title, xlabel, ylabel, limit=None):
    ax.set_title(title, fontsize=14)
    ax.set_xlabel(xlabel, fontsize=12)
    ax.set_ylabel(ylabel, fontsize=12)
    ax.grid(True, linestyle=':', alpha=0.6)

    for spine in ['top', 'bottom', 'left', 'right']:
        ax.spines[spine].set_color('black')
        ax.spines[spine].set_linewidth(1.5)

    if limit is not None:
        ax.set_xlim(0, limit)
        ax.set_ylim(0, limit)
        ax.set_aspect('equal', adjustable='box')


# --- Plot 1: (Ca+Mg) vs (HCO₃ + SO₄) in meq/L (1:1 Line) ---
fig, ax1 = plt.subplots(figsize=(7, 7))

# Calculate max limit for the 1:1 line
max_val_1 = max(df['Cations_Sum'].max(), df['Anions_Sum'].max()) * 1.1
ratio_coords_1 = [0, max_val_1]

ax1.scatter(df['Cations_Sum'], df['Anions_Sum'], label='Groundwater Samples', color='green', s=50)
ax1.plot(ratio_coords_1, ratio_coords_1, linestyle='--', color='black', linewidth=1.5, label='1:1 Ratio Line')

set_plot_style(ax1,
               'Cations vs Anions Partial Balance',
               r'$\mathrm{Ca^{2+}} + \mathrm{Mg^{2+}} \ (\mathrm{meq/L})$',
               r'$\mathrm{HCO_3^-} + \mathrm{SO_4^{2-}} \ (\mathrm{meq/L})$',
               limit=max_val_1)
ax1.legend(loc='upper left')
plt.tight_layout()
plt.show()


# --- Plot 2: K⁺ vs Cl⁻ in meq/L (1:1 Line) ---
fig, ax2 = plt.subplots(figsize=(7, 7))

# Calculate max limit
max_val_2 = max(df['K_meq'].max(), df['Cl_meq'].max()) * 1.1
ratio_coords_2 = [0, max_val_2]

ax2.scatter(df['K_meq'], df['Cl_meq'], label='Groundwater Samples', color='red', s=50)
ax2.plot(ratio_coords_2, ratio_coords_2, linestyle='--', color='gray', linewidth=1.5, label='1:1 Ratio Line')

set_plot_style(ax2,
               'K⁺ vs Cl⁻ Concentration (meq/L)',
               r'$\mathrm{K^{+}} \ (\mathrm{meq/L})$',
               r'$\mathrm{Cl^{-}} \ (\mathrm{meq/L})$',
               limit=max_val_2)
ax2.legend(loc='upper left')
plt.tight_layout()
plt.show()


# --- Plot 3: EC vs TDS (Linear Trend Line) ---
fig, ax3 = plt.subplots(figsize=(7, 7))

# Note: Using the standardized 'ec' and 'tds' columns
if 'ec' in df.columns and 'tds' in df.columns:
    # Calculate linear regression (TDS = k * EC + b)
    slope, intercept, r_value, p_value, std_err = linregress(df['ec'], df['tds'])
    r_squared = r_value**2

    # Generate the best-fit line data points
    p = np.poly1d([slope, intercept])
    x_fit = np.linspace(df['ec'].min(), df['ec'].max(), 100)

    # Scatter plot
    ax3.scatter(df['ec'], df['tds'], label='Groundwater Samples', color='purple', s=50)

    # Regression line
    line_label = f'$TDS = {slope:.2f} \cdot EC + {intercept:.2f} \quad (R^2 = {r_squared:.3f})$'
    ax3.plot(x_fit, p(x_fit), color='black', linestyle='-', label=line_label)

    set_plot_style(ax3,
                   'EC vs TDS Correlation',
                   r'Electrical Conductivity ($\mathrm{\mu S/cm}$ or $\mathrm{mS/cm}$)',
                   r'Total Dissolved Solids ($\mathrm{mg/L}$)')
    ax3.legend(loc='lower right')
else:
    ax3.text(0.5, 0.5, 'EC or TDS data columns not found.', ha='center', va='center', color='red')
    set_plot_style(ax3, 'EC vs TDS Correlation', 'EC', 'TDS')

plt.tight_layout()
plt.show()

print("All three plots have been successfully generated.")
```
